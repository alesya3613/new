```python
import pandas as pd

class BostonHousing:
    
    def __init__(self, file_way: str) -> None:
        self.df = pd.read_csv(file_way)
        
    def base_func(self, func: str) -> None:
        if func == 'info':
            return self.df.info()
        
        elif func == 'corr':
            return self.df.corr()
        
        elif func == 'head':
            return self.df.head()
        
    def unic_d(self, col: str) -> None:
        return self.df[col].unique()
    
    def graph_all(self, x_val: str, y_val: str, kind: str) -> None:
        return self.df.plot(x = x_val, y = y_val, kind = kind)
    
    def graph_mean(self, x_val: str, y_val: str, kind: str) -> None:
        self.df[x_val] = self.df[x_val].round(-1)
        group_data = self.df.groupby(x_val)[y_val].mean()
        return group_data.plot(x = x_val, y = y_val, kind = kind)
            
    
    
```

```python
data = BostonHousing('D:/BostonHousing.csv')
data.base_func('info')
```

```python
data.base_func('corr')
```

```python
data.base_func('head')
```

```python
data.graph_all('medv', 'rm', 'scatter')
```

```python
data.graph_mean('medv', 'rm', 'bar')
```

По графику можно наблюдать, что в общем с ростом количества комнат растет и цена на жилье, то есть существует прямая зависимость, за исключением некоторых случаев, где на цену в большей степени влияют другие факторы

```python
data.graph_all('medv', 'indus', 'scatter')
```

```python
data.graph_mean('medv', 'indus', 'bar')
```

```python
data.graph_all('medv', 'lstat', 'scatter')
```

```python
data.graph_mean('medv', 'lstat', 'bar')
```

В более дорогих домах процент населения с более низким статусом ниже, чем в более дешевых домах. Наблюдается обратная зависимость  

```python
data.graph_mean('medv', 'crim', 'bar')
```

```python
data.graph_all('medv', 'crim', 'scatter')
```

В основном в районах с большим процентом криминала дома стоят дешевле

```python
data.graph_mean('medv', 'ptratio', 'bar')
```

хотя зависимость между соотношением учеников и учителей и ценой дома довольно большая ~0,5, по графику видно, что оно не сильно колеблется в зависомости от цены, поэтому скорее всего влияет не так сильно
