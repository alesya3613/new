```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
iris = load_iris(as_frame=True)
```

```python
df = iris.frame
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f', cbar_kws={'label': ''})
plt.show()
```

```python
df['sepal length (cm)'].hist(bins=30, edgecolor='black',figsize=(10, 5))
plt.title('sepal length')
plt.show()
```

```python
df['sepal width (cm)'].hist(bins=30, edgecolor='black',figsize=(10, 5))
plt.title('sepal width (cm) ')
plt.show()
```

```python
df['petal length (cm)'].hist(bins=30, edgecolor='black',figsize=(10, 5))
plt.title('petal length (cm)')
plt.show()
```

```python
df['petal width (cm)'].hist(bins=30, edgecolor='black',figsize=(10, 5))
plt.title('petal width (cm)')
plt.show()
```
